```python
# importing libraries
import pandas as pd
import numpy as np
import os
```

```python
# DEFINING THE PATH

Path  = r'/Users/ghalebju/Desktop/Data Analysis Format/part 6/Zomato rest/Zomato Bangalore Restaurants analysis”/02 Data'
```

```python
# IMPORTING DATA FRAME

df = pd.read_csv(os.path.join (Path,'original Data','zomato.csv'))
```

```python
#exploring data set
df.info()
```

```python
#exploring data set
df.head()
```

```python
#dropping column "url"
df = df.drop(columns = ['url'])
```

```python
##exploring variables data type
df.dtypes
```

```python
#counting "nan" values in column "approx_cost"
(df['approx_cost(for two people)']=='nan').sum()
```

```python
#dropping Nan values
df = df.dropna()
```

```python
#deleting the comma from "approx_cost (for two people)" variable
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(',','')
```

```python
#changing data type for column "approx_cost"
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype('int64')
```

```python
#checking result
df.dtypes
```

```python
#basic statistical calculations for quantitative variables
df.describe()
```

```python
#deleting the "/5" from values in column "rate"
df['rate']=df['rate'].str.replace('/5','')
```

```python
df.head()
```

```python
#replacing the "." with "," for values in ccolumn "rate"
df['rate']=df['rate'].str.replace('.',',')
```

```python
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```
