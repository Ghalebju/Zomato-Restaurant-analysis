```python
# importing libraries
import pandas as pd
import numpy as np
import os
```

```python
# DEFINING THE PATH

Path  = r'/Users/ghalebju/Desktop/Data Analysis Format/Part 6'
```

```python
#importing data set

df = pd.read_csv(os.path.join(Path, 'zomato-last-version','01 Data','Original data' , 'zomato.csv'), index_col = False)
```

```python
#exploring data set
df.info()
```

```python
#exploring data set
df.head()
```

```python
#dropping column "url"
df = df.drop(columns = ['url'])
```

```python
#exploring variables data type
df.dtypes
```

```python
#deleting the comma from "approx_cost (for two people)" variable
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(',','')
```

```python
#counting "nan" values in column "approx_cost"
(df['approx_cost(for two people)']=='nan').sum()
```

```python
#dropping Nan values
df = df.dropna()
```

```python
#changing data type for column "approx_cost"
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype('int64')
```

```python
#checking result
df.dtypes
```

```python
#deleting the "/5" from values in column "rate"
df['rate']=df['rate'].str.replace('/5','')
```

```python
df = df[df['rate'] != 'NEW']
```

```python
df['rate'] = df['rate'].astype('float64')
```

```python
#checking result
df.dtypes
```

```python
#basic statistical calculations for quantitative variables
df.describe()
```

```python
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
#cheecking for missing values
df.isnull().sum()

#NO MISSING VALUES
```

```python
#exploting number of rows before duplicates
df.shape
```

```python
#creating subset with only dupliates
df_dups = df[df.duplicated()]
```

```python
#deleting duplicates
df_clean = df.drop_duplicates()
```
